### Lidando com valores faltantes

Não raro datasets possuem ruído, alguma lacuna que pode ser derivada tanto de um problema no processo de coleta de dados como a inexistencia da informação para determinada observação. 

Nos datasets esses casos podem ser representados como `?`, `NA`, `0`, ou mesmo uma _celula vazia_.  

O que se segue nesses casos é uma decisão do analista que pode ser simplesmente ignorar o problema até o extremo de eliminar a observação do dadtaset

Da decisão mais para menos drástica, temos 

 * o método `dropna` que elimina a observação do dataset
